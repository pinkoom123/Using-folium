```python
# Importing necessary libraries needed
import folium
from IPython.display import HTML
import pandas as pd
import numpy as np
```

```python
# reading and exploring data,hence dropping null data
M1=pd.read_csv('M1.csv',parse_dates=['Date'],index_col='Date')
columns_to_drop = [f'Unnamed: {i}' for i in range(7, 1025)]
M1 = M1.drop(columns=columns_to_drop, axis=1)
M1

```

```python
# Summing all empty or null columns

M1.isnull().sum()
```

```python
 

# Calculate the mean 
mean_values = np.mean(M1, axis=0)

print("Mean values for each column:")
print(mean_values)
```

```python

# Getting the map ready, using the coordintes

ghana_map = folium.Map(location=[7.25, -1.00], zoom_start=7)
```

```python
# Setting rainfall data according to the various coordinates(Lat & Lon) of the stations
# mean rainfall for various stations used
rainfall_data = {
'Axim': {'coordinates': [4.8699 , -2.2405],'rainfall': 90.895,},
'Tema':{'coordinates':[ 5.66980000, -0.01657000],'rainfall':77.059},
'Oda':{'coordinates':[5.928746,-0.974135],'rainfall':75.920},
'Ada':{'coordinates':[5.784521,0.649163], 'rainfall': 70.007},
'Saltpond':{'coordinates':[ 5.20913000,-1.06058000],'rainfall':82.068},
'Kumasi':{'coordinates':[ 6.6885, -1.6244],'rainfall':84.713}
}
```

```python
# For each region use the coordinates and rainfall value
for region, info in rainfall_data.items():
    folium.Marker(
        location=info['coordinates'],
        popup=f"{region} - Rainfall: {info['rainfall']} mm",
    ).add_to(ghana_map)
```
